```python
# object columns
obj=cancer.select_dtypes(include=['object'])
num=cancer.select_dtypes(include=np.number)
```

```python
obj.columns
```

```python
num.columns
```

```python
# LabelEncode
from sklearn.preprocessing import LabelEncoder
```

```python
objdummy=obj.apply(LabelEncoder().fit_transform)
```

```python
objdummy.head()
```

```python
df=pd.concat([objdummy,num],axis=1)
df.head()
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
y=df['diagnosis']
X=df.drop(['diagnosis','id'],axis=1)
```

```python
le=LogisticRegression(max_iter=5000)
```

```python
lemodel=le.fit(X,y)
```

```python
lemodel.score(X,y)
```

```python
lemodel.intercept_
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lemodel,X,y,cv=5)
```

```python
np.mean([0.93859649, 0.94736842, 0.98245614, 0.92982456, 0.96460177])
```

```python
pd.DataFrame(lemodel.coef_,columns=X.columns)
```

```python
lepredict=lemodel.predict(X)
```

```python
pd.crosstab(y,lepredict)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y,lepredict))
```

```python
from sklearn.metrics import plot_roc_curve
```

```python
plot_roc_curve(le,X,y)
```
